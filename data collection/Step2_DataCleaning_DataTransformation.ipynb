## Data cleaning

```python
listings[listings.reviewnumber == '[]']
```

```python
listings = listings[listings.reviewnumber != '[]']
listings.shape
```

```python
#Remove emoji
listings["title"]=listings["title"].str.replace('[^\w\s+&#@/:%.,_-]', '', flags=re.UNICODE)
```

```python
# transform reviwnumber attribute in an integer attribute
for i in range(listings.shape[0]):
    
    num = int(re.findall(r"[0123456789]+",listings.iloc[i]["reviewnumber"])[0])
    listings.loc[i,"reviewnumber"] = num
    
```

```python
listings.dropna(subset = ["listing_id"], inplace=True)
listings['listing_id'] = listings['listing_id'].apply(np.int64)
```

```python
grouped = binetwork.groupby(binetwork.date)
snapshot2009 = grouped.get_group('2009') 
snapshot2010 = grouped.get_group('2010') 
snapshot2011 = grouped.get_group('2011') 
snapshot2012 = grouped.get_group('2012') 
snapshot2013 = grouped.get_group('2013')
snapshot2014 = grouped.get_group('2014') 
snapshot2015 = grouped.get_group('2015') 
snapshot2016 = grouped.get_group('2016') 
snapshot2017 = grouped.get_group('2017')
snapshot2018 = grouped.get_group('2018')
snapshot2019 = grouped.get_group('2019') 
snapshot2020 = grouped.get_group('2020') 
snapshot2021 = grouped.get_group('2021') 
```

```python
snapshot2020.head()
```

```python
def weightsnap(snap_df):
    snap_df = snap_df.groupby(['reviewer_id','listing_id'])[['date']].count().reset_index()
    return snap_df.rename(columns={'date': 'weight'}) 
```

```python
snapshot2009 = weightsnap(snapshot2009)
snapshot2010 = weightsnap(snapshot2010)
snapshot2011 = weightsnap(snapshot2011)
snapshot2012 = weightsnap(snapshot2012)
snapshot2013 = weightsnap(snapshot2013)
snapshot2014 = weightsnap(snapshot2014)
snapshot2015 = weightsnap(snapshot2015)
snapshot2016 = weightsnap(snapshot2016)
snapshot2017 = weightsnap(snapshot2017)
snapshot2018 = weightsnap(snapshot2018)
snapshot2019 = weightsnap(snapshot2019)
snapshot2020 = weightsnap(snapshot2020)
snapshot2021 = weightsnap(snapshot2021)
```
